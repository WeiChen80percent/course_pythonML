```python
def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\pi/2$"
    elif N == 2:
        return r"$\pi$"
    elif N % 2 > 0:
        return r"${0}\pi/2$".format(N)
    else:
        return r"${0}\pi$".format(N // 2)

ax = plt.axes()
ax.grid()
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.cos(x)
ax.plot(x, y)
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
```

### 3. Custom Tick Locations
Plot a sine curve and customize both x and y ticks to show specific values using `plt.xticks()` and `plt.yticks()`.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 100)
x_tick = np.arange(-2*np.pi, 2*np.pi+1, np.pi/2)
y = np.sin(x)
y_tick = np.arange(-1, 1.5, 0.5)
plt.plot(x, y)
plt.xticks(x_tick, rotation=45)
plt.yticks(y_tick, rotation=45)
```

### 8. Basic 3D Line Plot

### 9. 3D Surface Plot with a Color Map
Generate a 3D surface plot of a mathematical function (e.g., `z = sin(x) * cos(y)`) using `ax.plot_surface()`. Apply a colormap like `'viridis'` or `'plasma'`.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax = plt.axes(projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x,y = np.meshgrid(x,y)
z = np.sin(x) * np.cos(y)
ax.plot_surface(x,y,z, cmap='viridis')
```

### 10. Wireframe vs Surface Comparison
Create a figure with two subplots: one showing a wireframe plot and the other showing a surface plot of the same function. Use titles and consistent view angles for comparison.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
fig, ax = plt.subplots(1,2, figsize=(10, 10),subplot_kw={'projection':'3d'})
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x,y = np.meshgrid(x,y)
z = np.sin(x) + np.cos(y)
ax[0].plot_surface(x,y,z, cmap='viridis')
ax[0].set_title("surface")
ax[1].plot_wireframe(x,y,z, color='red')
ax[1].set_title("wireframe")

```
