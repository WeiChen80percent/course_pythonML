### **Exercise 5: Array Indexing and Slicing**
Access and modify elements in a NumPy array.


### **Exercise 9**: Aggregations with ufuncs
Use NumPys aggregation functions.

Instructions:

1. Create a NumPy array of numbers from `1` to `10`.
2. Compute and print:
    - The sum of all elements.
    - The mean (average) of all elements.
    - The maximum and minimum values in the array.
    - The product of all elements.

````python
arr = []
for i in range(1,11):
    arr.append(i)

arr = np.array(arr)
arrsum = np.sum(arr)
print(arrsum)

arrmean = np.mean(arr)
print(arrmean)

arrmax = np.max(arr)
print(arrmax)

arrmin = np.min(arr)
print(arrmin)

arrprod = np.prod(arr)
print(arrprod)
````
