## [Simple Line Plots](https://jakevdp.github.io/PythonDataScienceHandbook/04.01-simple-line-plots.html)

Use `plt.plot(xs, ys)` to draw segements.  
You may change the **colors** and line **styles**.  
You may **adjust each axis** and **add labels**.  

Check out details by `plt.plot?`.

```python
x = np.linspace(0, 10, 1000)
```

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
# tight, equal, off
```

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

- `plt.xlabel()`  `ax.set_xlabel()`

- `plt.ylim()`  `ax.set_ylim()`

## Exercise 9: Scatter Plot with Color Mapping
Generate a scatter plot where each points color depends on a third variable (e.g., distance from the origin). Use `c=` in `plt.scatter()` and include a colorbar.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.normal(0,30,1000)
y = np.random.normal(0,30,1000)
cluster = lambda x,y: np.int16(np.sqrt(x**2+y**2)) 
plt.scatter(x,y,s=50,c=cluster(x,y),alpha=0.5)
plt.colorbar()

plt.show()
```

## Exercise 10: Multiple Subplots
Create a figure with 2x2 subplots showing: a line plot, scatter plot, histogram, and contour plot. Use `plt.subplots()` and organize the layout with proper titles.

```python
import numpy as np
import matplotlib.pyplot as plt

fig,axs = plt.subplots(2,2)

x = np.linspace(-10,11,100)
y = np.random.normal(0,0.1,100)

axs[0][0].plot(x,y,color="black")
axs[0][0].set_title("line plot")

axs[0][1].scatter(x,y,s=50,color='purple',alpha=0.5) 
axs[0][1].set_title("scatter plot")

axs[1][0].hist2d(x,y,bins=30,cmap='Blues')
axs[1][0].set_title("hist2d")

z = np.sin(x**2+y[:,np.newaxis]**2)
axs[1][1].contour(x,y,z)
axs[1][1].set_title("contour")

plt.tight_layout()
plt.show()
```
